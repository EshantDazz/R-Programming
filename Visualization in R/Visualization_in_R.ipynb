### `Aesthetics is something with what our data maps with something which we wanna see`



```r
head(df,n=2)
```

```r
colnames(df)<-c("Film","Genre","CriticalRating","AudienceRating","BudgetMillions","Year")
```

```r
head(df,n=2)
```

```r
ggplot(data=df,aes(x=CriticalRating,y=AudienceRating))
```

```r
##Adding Geometry
ggplot(data=df,aes(x=CriticalRating,y=AudienceRating)) + geom_point()
```

```r
##Adding Genre
##Adding Geometry
ggplot(data=df,aes(x=CriticalRating,y=AudienceRating,color=Genre,size=BudgetMillions)) + geom_point()
```

### `Working with Layers`

```r
p<-ggplot(data=df,aes(x=CriticalRating,y=AudienceRating,color=Genre,size=BudgetMillions)) 
```

```r
p
```

```r
p+geom_point()
```

```r
p+geom_line()
```

```r
p+geom_line()+geom_point()
```

### `Overiding Austhetics`

```r
q<-ggplot(data=df,aes(x=CriticalRating,y=AudienceRating,color=Genre,size=BudgetMillions)) 
```

```r
#Adding Layers
q+geom_point()
```

```r
##Overiding aes
#Ex 1
q+geom_point(aes(size=CriticalRating))
```

```r
##Ex2
q+geom_point(aes(color=BudgetMillions))
```

### `Mapping vs Setting`

```r
r<-ggplot(data=df,aes(x=CriticalRating,y=AudienceRating)) 
```

```r
r+geom_point()
```

### Adding colors
- Mapping
- Setting

```r
#Mapping
r+geom_point(aes(color=Genre))
```

```r
#Setting
r+geom_point(color="DarkGreen")
```
