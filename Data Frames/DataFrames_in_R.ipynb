## <font color='blue'> Merging DataFrames</font>

```r
a<-c(1,2,3,4,5)
b<-c('eshant','das','hello','papu','hehe')
cc<-c('help',2,43,5,'bye')
```

```r
e<-c(54,42,136,46,521)
f<-c('lol','lets go','happy','hmm',2)
```

```r
d1<-data.frame(a,b,cc)
```

```r
d2<-data.frame(pop=a,lop=b,dop=cc)
```

```r
d1
```

```r
d2
```

```r
merged<-merge(d1,d2,by.x="b",by.y="lop")
```

```r
merged
```
